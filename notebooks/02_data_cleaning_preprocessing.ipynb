```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_excel("Sample - Superstore.xls")

df.head()
```

```python
df.isnull().sum()
```

```python
df["Postal Code"] = df["Postal Code"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["Postal Code"] = df["Postal Code"].astype(int)
```

```python
df["Postal Code"].dtype
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("/", "_")
)
```

```python
df.columns.tolist()
```

```python
df["order_year"] = df["order_date"].dt.year

df["order_month"] = df["order_date"].dt.month

df["order_month_name"] = df["order_date"].dt.month_name()

df["order_quarter"] = df["order_date"].dt.quarter
```

```python
df[
    [
        "order_date",
        "order_year",
        "order_month",
        "order_month_name",
        "order_quarter"
    ]
].head()
```

```python
df["shipping_days"] = (
    df["ship_date"] - df["order_date"]
).dt.days
```

```python
df["shipping_days"].describe()
```

```python
df["profit_margin_pct"] = (
    df["profit"] / df["sales"]
) * 100
```

```python
df["profit_margin_pct"] = (
    df["profit_margin_pct"]
      .round(2)
)
```

```python
df[
    [
        "sales",
        "profit",
        "profit_margin_pct"
    ]
].head()
```

```python
loss_orders = (
    df["profit"] < 0
).sum()

print("Loss Making Orders:", loss_orders)
```
